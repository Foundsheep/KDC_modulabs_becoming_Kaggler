```python
# print("Before removal :", len(train))

# train.loc[(train['grade']==7) & (train['price']>1500000)] # 2개
# train.loc[(train['grade']==8) & (train['price']>2000000)] # 3개
# train.loc[(train['grade']==10) & (train['price']>3100000)] # 2개
# train.loc[(train['grade']==11) & (train['price']>3400000)] # 4개

# print("After removal :", len(train))

index = []
conditions =[(train['grade']==7) & (train['price']>1500000),
           (train['grade']==8) & (train['price']>2000000),
           (train['grade']==10) & (train['price']>3100000),
           (train['grade']==11) & (train['price']>3400000)]

for condition in conditions:
    for indx in train.loc[condition].index:
        index.append(indx)
print(index)
```

```python
print("Before removal :", len(train))
print("Numbers of removal :", len(index))

train.drop(index=index, axis=0, inplace=True)

print("After removal :", len(train))
```

```python
# 한 곳에 모아보기

# GridSearch 가져 오기
from sklearn.model_selection import GridSearchCV
print("loaded")

# GridSearch를 위한 XGB 모델 만들기
model = xgb.XGBRegressor()

# 하이퍼파라미터 리스트 설정
params = {'booster':['gblinear'],
         'nthread':[1,10,20],
         'learning_rate':[0.0001,0.001,0.01,0.1,1],
         'min_child_weight':[1,2,3,4]}

# GridSearch 모델 학습
grid_model = GridSearchCV(model, param_grid=params, scoring='neg_mean_squared_error', cv=5, verbose=1, n_jobs=5
                         )
grid_model.fit(train,y)

# 결과 확인 및 정리
grid_model.cv_results_
param_results = grid_model.cv_results_['params']
score_results = grid_model.cv_results_['mean_test_score']
results = pd.DataFrame(param_results)
results['score'] = score_results
RMSE = np.sqrt((results['score'] * -1))
results['RMSE'] = RMSE
results.sort_values(by='RMSE', axis=0)

# XGB 모델에 최적 파라미터 넣기
model = xgb.XGBRegressor(learning_rate=1, min_child_weight=3, nthread=10)
model.fit(train, y)
y_pred = model.predict(test)
y_pred
```
